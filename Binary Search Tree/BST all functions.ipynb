```python
#Code for BST with all the functions- insert, delete, print, count , min

class BinaryTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
class BST:
    def __init__(self):
        self.root = None
        self.numNodes = 0
        
    def count(self):
        return self.numNodes
        
    def printTreeHelper(self,root):
        if root == None:
            return
        print(root.data, end=":")
        if root.left != None:
            print("L", root.left.data, end='')
        if root.right != None:
            print("R", root.left.data, end='')   
        print()
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)
        
    def printTree(self):
        self.printTreeHelper(self.root)
    
    def isPresentHelper(self,root,data):
        if root == None:
            return False
        
        if root.data == data:
            return True
        
        if root.data > data:
            #call on left
            return self.isPresentHelper(root.left,data)
        else:
            #call on right
            return self.isPresentHelper(root.right,data)
    
    def isPresent(self,data):
        return self.isPresentHelper(self.root,data)
    
    def insertHelper(self,root,data):
        
        #if the tree is empty, create a node and return it
        if root == None:
            node = BinaryTreeNode(data)
            return node
        
        #if data is left, insert data on the left side and connect it to root.left
        if root.data > data:
            root.left = self.insertHelper(root.left,data)
            return root
        
        #repeat the same for the right side and connect it to root.right
        else:
            root.right = self.insertHelper(root.right,data)
            return root
        
    def insert(self,data):
        self.numNodes += 1
        self.root = self.insertHelper(self.root,data)
        
    
    def min(self,root):
        
        #to find minimum of an empty, return a large number
        if root == None:
            return 10000
        
        #if left is Nonem then root is the minimum
        if root.left == None:
            return root.data
        
        #if left is not None, then minimum on left side only
        #NOTE: in a BST, minimum can only be found on LHS of BST.
        return self.min(root.left)
    
    def deleteDataHelper(self,root,data):
        
        #empty Tree
        if root == None:
            return False, None
        
        if root.data < data:
            deleted, newRightNode = self.deleteDataHelper(root.right,data)
            root.right = newRightNode
            return deleted, root
        
        if root.data > data:
            deleted, newLeftNode = self.deleteDataHelper(root.left,data)
            root.left = newLeftNode
            return deleted, root
        
        #leaf node
        if root.left == None and root.right == None:
            return True, None
        
        #one child
        #contains right child only
        if root.left == None:
            return True, root.right
        
        #contains left child only
        if root.right == None:
            return True, root.left
        
        #root has two children:
        replacement = self.min(root.right)
        root.data = replacement
        deleted, newRightNode = self.deleteDataHelper(root.right, replacement)
        root.right = newRightNode
        return True, root
    
    def deleteData(self,data):
        
        deleted,newRoot = self.deleteDataHelper(self.root,data)
        if deleted:
            self.numNodes -= 1
        self.root = newRoot
        return deleted

b = BST()
b.insert(10)
b.insert(5)
b.insert(12)
print(b.isPresent(10))
print(b.deleteData(10))
print(b.count())
b.printTree()
```
