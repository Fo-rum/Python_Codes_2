Given an integer array (of length n), find and return all the subsets of input array.

Subsets are of length varying from 0 to n, that contain elements of the array. 

But the order of elements should remain same as in the input array.

Note : The order of subsets are not important.

Input format :

Line 1 : Size of array

Line 2 : Array elements (separated by space)

```python
## Read input as specified in the question.
## Print output as specified in the question.

#my code
def subsetofarray(arr):
    
    #base case:
    #output is 2-
    if len(arr) == 0:
        return [[]]
        
    #induction hypothesis
    small_output = subsetofarray(arr[1:])
    
    new_length = 2 * len(small_output)
    output = new_length * [0]
    
    index = 0
    
    #copy the elements of the old array in the new array
    for i in range(len(small_output)):
        output[index] = small_output[i]
        index += 1 
        
    for i in range(len(small_output)):
        
        #getting the length of the small output 
        new_length = len(small_output[i]) + 1
        
        #initializing all zeros
        output[index] = new_length * [0]
        
        #putting the first element of array in its respective position
        output[index][0] = arr[0]
        
        #copy the elements of the small_output in the output array after the first element is inserted in its respective position
        
        #jagged array concept used
        for j in range(len(small_output[i])):
            output[index][j+1] = small_output[i][j]
        index += 1
    
    return output



n = int(input())
arr = [int(ele) for ele in input().split()]
ans = subsetofarray(arr)
for i in ans:
    print(*i)
    
#not my code
# def subsetOfArray(arr):
    
#     #base case
#     if len(arr)==0:
#         return [[]]
    
#     smallOutput=subsetOfArray(arr[1:])
#     output=(2*len(smallOutput))*[0]
#     index=0
    
#     for i in range(len(smallOutput)):
#         output[index]=smallOutput[i]
#         index+=1
    
#     for i in range(len(smallOutput)):
        
#         #here we are filling each array which is initially 0 with smallout[i]
#         output[index]=(len(smallOutput[i])+1)*[0]
#         output[index][0]=arr[0]
        
# 		# it means that at each index we are assigning a list as an array element
#         for j in range(len(smallOutput[i])):
#             output[index][j+1]=smallOutput[i][j]
#         index+=1
#     return output
        
        

# n=int(input())
# arr=[int(ele) for ele in input().split()]
# totalsubset=subsetOfArray(arr)
# for ele in totalsubset:
#     print(*ele)
```
