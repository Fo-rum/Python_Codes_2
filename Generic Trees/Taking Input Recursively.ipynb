```python
"""Induction step:
1. taking the root input
2. asking for the number of children of root
3. making those many recursive calls
4. relevant connections made
"""
```

```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.children = []
        
def printTree(root):
    
    #edge case, not a base case:
    if root == None:
        return
    print(root.data)
    for child in root.children:
        printTree(child)
        
def printTreeDetailed(root):
    
    #edge case
    if root == None:
        return
    print(root.data, ":", end='')
    
    #for loop for printing all the children of the root
    for child in root.children:
        print(child.data, "," , end="")
    print()
    
    #for loop for making a function call on each and every child of the root
    for child in root.children:
        printTreeDetailed(child)
        
def takeTreeInputR():
    print("Enter root data:")
    rootData = int(input())
    
    #edge case - incase the user doesnt want to pass anything
    if rootData == -1:
        return None
    
    root = TreeNode(rootData)
    print("Enter the number of children for ", rootData)
    childrenCount = int(input())
    
    for i in range(childrenCount):
        child = takeTreeInputR()
        root.children.append(child)
        
    return root

root = takeTreeInputR()
print("tree details:")
printTreeDetailed(root)
                           
```
