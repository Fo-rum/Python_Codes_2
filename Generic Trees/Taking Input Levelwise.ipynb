```python
import queue

class TreeNode:
    def __init__(self,data):
        self.data = data
        self.children = []
        
def printTree(root):
    
    #edge case, not a base case:
    if root == None:
        return
    print(root.data)
    for child in root.children:
        printTree(child)
        
def printTreeDetailed(root):
    
    #edge case
    if root == None:
        return
    print(root.data, ":", end='')
    
    #for loop for printing all the children of the root
    for child in root.children:
        print(child.data, "," , end="")
    print()
    
    #for loop for making a function call on each and every child of the root
    for child in root.children:
        printTreeDetailed(child)
        
def takeTreeInputLevelwise():
    q = queue.Queue()
    print("Enter root:")
    rootData = int(input())
    if rootData == -1:
        return None
    
    root = TreeNode(rootData)
    q.put(root)
    while(not(q.empty())):
        current_node = q.get()
        print("Enter the no of children for", current_node.data)
        numChildren = int(input())
        for i in range(numChildren):
            print("Enter next child for", current_node.data)
            childData = int(input())
            child = TreeNode(childData)
            current_node.children.append(child)
            q.put(child)
    
    return root

root = takeTreeInputLevelwise()
print("tree details:")
printTreeDetailed(root)
```
