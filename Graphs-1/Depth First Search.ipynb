```python

class Graph:
    
    def __init__(self,nVertices):
        self.nVertices = nVertices
        
        #creating an adjacency matrix of size n*n
        self.adjMatrix = [[0 for j in range(nVertices)] for i in range(nVertices)]
        
    def addEdge(self,v1,v2):
        
        #undirected graph
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1
    
    def removeEdge(self,v1,v2):
        
        #undirected graph
        if self.containsEdge(v1,v2) is False:
            return
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    def containsEdge(self,v1,v2):
        if self.adjMatrix[v1][v2] > 0:
            return True
        else:
            return False
        
    def __dfsHelper(self,sv,visited):
        print(sv) #sv means starting vertex
        visited[sv] = True
        for i in range(self.nVertices):
            if (self.adjMatrix[sv][i] > 0 and visited[i] is False):
                self.__dfsHelper(i,visited)
    
    def dfs(self):
        visited = [False for i in range(self.nVertices)]
        self.__dfsHelper(0,visited)
    
    def __str__(self):
        #overriding the default print method. Without this, print(g) would print the graph object location
        return str(self.adjMatrix)         # typecasting the adjacency matrix as str
    
g = Graph(6)
g.addEdge(0,1)
g.addEdge(1,5)
g.addEdge(5,2)
g.addEdge(2,3)
g.addEdge(2,4)
g.addEdge(0,2)

g.dfs()
```

```python
#the above algorithm wont work for disconnected graphs.
#That is it will only check for one graph and return
# so now we have to see that after visiting one graph, if something isnt visited, then we have to visit the remaining vertices.

#Making connections in the above dfs code for disconnected graphs.

class Graph:
    
    def __init__(self,nVertices):
        self.nVertices = nVertices
        
        #creating an adjacency matrix of size n*n
        self.adjMatrix = [[0 for j in range(nVertices)] for i in range(nVertices)]
        
    def addEdge(self,v1,v2):
        
        #undirected graph
        self.adjMatrix[v1][v2] = 1
        self.adjMatrix[v2][v1] = 1
    
    def removeEdge(self,v1,v2):
        
        #undirected graph
        if self.containsEdge(v1,v2) is False:
            return
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    def containsEdge(self,v1,v2):
        if self.adjMatrix[v1][v2] > 0:
            return True
        else:
            return False
        
    def __dfsHelper(self,sv,visited):
        print(sv) #sv means starting vertex
        visited[sv] = True
        for i in range(self.nVertices):
            if (self.adjMatrix[sv][i] > 0 and visited[i] is False):
                self.__dfsHelper(i,visited)
    
    def dfs(self):
        visited = [False for i in range(self.nVertices)]
        
        #first it starts from 0. then checks for 1 which is visited via 0
        # So, then it starts from 2 and applies dfs.
        for i in range(self.nVertices):
            if visited[i] is False:
                self.__dfsHelper(i,visited)
    
    def __str__(self):
        #overriding the default print method. Without this, print(g) would print the graph object location
        return str(self.adjMatrix)         # typecasting the adjacency matrix as str
    
g = Graph(7)
#graph-1 vertices
g.addEdge(0,1)
g.addEdge(0,3)

#graph-2 vertices
g.addEdge(2,4)
g.addEdge(2,5)
g.addEdge(4,6)

g.dfs()
```
